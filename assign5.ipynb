**.Apply method** is a method that applies a function along an axis of a dataframe, it is an alternative to using a loop to iterate over a dataframe. It takes a function as an input and applies it to the whole dataframe. 
The .apply() method helps to segregate data according to the conditions required, and it is efficiently used in data science and machine learning. 

When using the lambda function within the apply function, the lambda function is going to be applied to every element along the specified axis.

The general form of the .apply() method with the lambda function is:

df['new_column'] = df['existing_column'].apply(lambda x: <lambda_operation>)

Now i will create a custom function to use within an apply function on our previous dataset.

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\HP i5\Desktop\SE_Factory\Week 1\Pythonex\kc_house_data.csv")
```

```python
df.head()
```

```python
def categorize_price(price):
    if price <= 500000:
        return 'Low'
    elif price <= 1000000:
        return 'Moderate'
    else:
        return 'High'
```

```python
df['price_bin'] = df['price'].apply(categorize_price)
print(df[['price', 'price_bin']])
```

```python
df_ = df[['id', 'price', 'sqft_living', 'waterfront', 'view']].copy()
```

```python
df_['price_bin'] = df_.apply(lambda x:'High' if x['price'] > 1000000 else 'Medium' if x['price'] > 500000   else 'Low', axis = 1 )
df_.head(10)
```

```python
df_[['sqft_living', 'price_bin']].groupby('price_bin').mean()
```

Another custom function:

```python
df_['view_categ'] = df_.apply(lambda x: 'High view' if x['view']>=3 else 'Low View' if x['view']>=1 else 'No View', axis = 1)
df_.head()
```

```python
df_[['sqft_living', 'price', 'view_categ']].groupby('view_categ').mean()
```

Now lets try an example where we calculate the price per square feet of living space:

```python
df_['price_per_sqft'] = df_.apply(lambda x: x['price'] / x['sqft_living'], axis=1)
df_.head()
```
